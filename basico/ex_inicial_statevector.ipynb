```python
from qiskit import *
from qiskit.visualization import plot_histogram
from math import *

qc = QuantumCircuit(1)
inicial_state = [1/sqrt(2),complex(0,1/sqrt(2))]
qc.initialize(inicial_state, 0)
# qc.measure_all() - ATENÇÃO: se fizer a medida o vetor estado colapsa
qc.draw()
```

```python
statevector = Aer.get_backend('statevector_simulator')
result = execute(qc,backend = statevector).result()
print(result.get_statevector())
```

```python
plot_histogram(result.get_counts(qc))
```

```python
qc = QuantumCircuit(1)
vector = [1/sqrt(3), sqrt(1 - 1/3)]
qc.initialize(vector,0)
qc.draw()
```

```python
statevector = Aer.get_backend('statevector_simulator')
result = execute(qc, statevector).result()
print(result.get_statevector())
```

```python
plot_histogram(result.get_counts())
```

```python
#Esfer de bloch
from qiskit.tools.visualization import plot_bloch_multivector
plot_bloch_multivector(result.get_statevector())
```
