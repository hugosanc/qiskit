```python
from qiskit import *
from math import *
from qiskit.visualization import plot_histogram

qc = QuantumCircuit(3)
for n in range(3):
  qc.h(n)
qc.draw('mpl')
```

```python
simulator = Aer.get_backend('statevector_simulator')
final_state = execute(qc,simulator).result()

print(final_state.get_statevector())
```

```python
plot_histogram(final_state.get_counts(qc))
# qc.draw()
```

```python
qc = QuantumCircuit(2,2)
qc.h(0)
qc.x(1)
for n in range(2):
    qc.measure(n,n)
qc.draw('mpl')
```

```python
provider = IBMQ.get_provider('ibm-q')
qcomp = provider.get_backend('ibmq_ourense')
job = execute(qc, qcomp)
from qiskit.tools.monitor import job_monitor
job_monitor(job)
```

```python
result = job.result()
plot_histogram(result.get_counts(qc))
```
