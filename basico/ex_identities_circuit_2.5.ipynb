# Swapping qubits

```python
from qiskit import *
from math import *
from qiskit.visualization import plot_histogram, plot_bloch_multivector

qc = QuantumCircuit(2)
# swap qubits a and b
a = 0
b = 1
qc.x(a)
qc.swap(a,b)
qc.draw('mpl')
```

```python
simulator = Aer.get_backend('statevector_simulator')
result = execute(qc,simulator).result()
plot_histogram(result.get_counts())
```

```python
qc = QuantumCircuit(2)
# swap qubits a and b
a = 0
b = 1
qc.h(a)
qc.cx(a,b)
qc.cx(b,a)
qc.draw('mpl')
```

```python
simulator = Aer.get_backend('statevector_simulator')
result = execute(qc,simulator).result()
plot_histogram(result.get_counts())
```

```python
simulator = Aer.get_backend('statevector_simulator')
result = execute(qc,simulator).result()
plot_histogram(result.get_counts())
```

## Controlled rotations

```python
qc = QuantumCircuit(2)
theta = pi # theta can be anything (pi chosen arbitrarily)
c = 0
t = 1
qc.h(c)
qc.ry(theta/2,t)
qc.cx(c,t)
qc.ry(-theta/2,t)
qc.cx(c,t)
qc.draw('mpl')
```

```python
simulator = Aer.get_backend('statevector_simulator')
result = execute(qc,simulator).result()
plot_histogram(result.get_counts())
```

```python
plot_bloch_multivector(result.get_statevector())
```

## Toffoli gate

```python
qc = QuantumCircuit(3)
a = 0
b = 1
t = 2
qc.x(a)
qc.x(b)
qc.x(t)
# Toffoli with control qubits a and b and target t
qc.ccx(a,b,t)
qc.draw('mpl')
```

```python
simulator = Aer.get_backend('statevector_simulator')
result = execute(qc,simulator).result()
plot_histogram(result.get_counts())
```

```python
qc = QuantumCircuit(1)
qc.h(0)
qc.t(0)
qc.h(0)
qc.t(0)
qc.draw('mpl')
```

```python
simulator = Aer.get_backend('statevector_simulator')
result = execute(qc,simulator).result()
plot_histogram(result.get_counts())
```

```python
plot_bloch_multivector(result.get_statevector())
```
