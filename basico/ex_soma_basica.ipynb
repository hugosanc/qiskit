```python
from qiskit import *
```

```python
qc = QuantumCircuit(4,2)
qc.x(0)
qc.x(1)

qc.barrier()
qc.cx(0,2)
qc.cx(1,2)
qc.ccx(0,1,3)

qc.barrier()
qc.measure(2,0)
qc.measure(3,1)
qc.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(qc,backend = simulator).result()
from qiskit.visualization import plot_histogram
plot_histogram(result.get_counts(qc))
```
